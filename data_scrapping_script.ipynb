### Infinite Scrolling

```python
SCROLL_PAUSE_TIME = 20

# Get scroll height
last_height = driver.execute_script("return document.body.scrollHeight")

while True:
    # Scroll down to bottom
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    # Wait to load page
    time.sleep(SCROLL_PAUSE_TIME)

    # Calculate new scroll height and compare with last scroll height
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height

container=[]
Product_Name_tag=driver.find_elements(By.XPATH,"//div[@class='_1l4fi']")
for k in Product_Name_tag:
    p=k.text.split("\n")
    container.append(p)
```

```python
SCROLL_PAUSE_TIME = 20

# Get scroll height
last_height = driver.execute_script("return document.body.scrollHeight")

while True:
    # Scroll down to bottom
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    # Wait to load page
    time.sleep(SCROLL_PAUSE_TIME)

    # Calculate new scroll height and compare with last scroll height
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height

container=[]
Product_Name_tag=driver.find_elements(By.XPATH,"//div[@class='_1l4fi']")
for k in Product_Name_tag:
    p=k.text.split("\n")
    container.append(p)
```

```python
Make_year=[]
Brand=[]
Model=[]
Transmission=[]
Kms=[]
Owner=[]
variant=[]
Price=[]
City=[]
```

```python
container1=[]
Product_Name_tag=driver.find_elements(By.XPATH,"//div[@class='_7udZZ']")
for k in Product_Name_tag:
    p=k.text.split("\n")
    container1.append(p)
    
for i in range(0,len(container1)):
    Price.append(container1[i])
```

```python
container2=[]
Product_Name_tag=driver.find_elements(By.XPATH,"//ul[@class='bVR0c']")
for k in Product_Name_tag:
    p=k.text.split("\n")
    container2.append(p)
```

```python
for i in range(0,len(container2)):
    if 'Hybrid' in container[i][1]:
        variant.append('Petrol + Ev')
    elif 'HYBRID' in container[i][1]:
        variant.append('Petrol + Ev')
    else:
        variant.append(container2[i][2])
    
for i in range(0,len(container)):
    a=container[i][0]
    Make_year.append(a[0:4])
    Brand.append(a[5::])
    Model.append(container[i][1])
    Kms.append(container[i][2])
    Owner.append(container[i][3])
    City.append("Hyderabad")
    if 'Automatic' in container[i][1]:
        Transmission.append('Automatic')
    elif 'AUTOMATIC' in container[i][1]:
        Transmission.append('Automatic')
    else:
        Transmission.append('Manual')
        
data=list(zip(Make_year,Brand,Model,Transmission,Kms,Owner,variant,City,Price))
              
df7= pd.DataFrame(data, columns=['Make_Year','Brand','Model','Transmission','Kms','Owner','Variant','City','Price'])

df7
```

```python
driver.quit()
```
